```python
# question 1
def comma(n):
    l = []
    for i in range(n+1):
        if i % 5 == 0 and i % 7 == 0:
            l.append(i)
    return l
def NumGen(n):
    a = comma(n)
    b = len(a)-1
    for i in range(0, n+1):
        if i % 5 == 0 and i % 7 == 0:
            yield i
            if i == a[b]:
                yield ""
            else:
                yield ','

            
n = int(input())

for i in NumGen(n):
    print(i, end = " ")
```
